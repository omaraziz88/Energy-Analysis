```python
import pandas as pd
```

```python
co2_emissions = "CO2_Emissions.csv"
crude_ngl_production = "Crude_NGL_Production.csv"
natural_gas_production="Natural_Gas_Production.csv"
renewables_electricity_share="Renewables_Electricity_Share.csv"
solar_wind_share="Solar_and_Wind_Share.csv"
electricity_production="Total_Electricity_Production.csv"
included_columns=['World','OECD','G7','BRICS','Europe','North America','Latin America','Pacific','Africa','Middle-East']
```

```python
co2_emissions_df = pd.read_csv(co2_emissions, encoding="ISO-8859-1")
co2_emissions_df.set_index('Country',inplace=True)
co2_emissions_df.head()
co2_emissions_df_transposed=co2_emissions_df.T
co2_emissions_df_transposed.head()
```

```python
co2_emissions_df_transposed.drop(co2_emissions_df_transposed.columns.difference(included_columns), 1, inplace=True)
co2_emissions_df_transposed.plot(kind='line')
```

```python
crude_ngl_production_df = pd.read_csv(crude_ngl_production, encoding="ISO-8859-1")
crude_ngl_production_df.set_index('Country',inplace=True)
crude_ngl_production_df.head()
crude_ngl_production_transposed=crude_ngl_production_df.T
crude_ngl_production_transposed.head()
```

```python
crude_ngl_production_transposed.drop(crude_ngl_production_transposed.columns.difference(included_columns), 1, inplace=True)
crude_ngl_production_transposed.plot(kind='line')
```

```python
natural_gas_production_df= pd.read_csv(natural_gas_production, encoding="ISO-8859-1")
natural_gas_production_df.head()
natural_gas_production_df.set_index('country',inplace=True)
natural_gas_production_transposed=natural_gas_production_df.T
natural_gas_production_transposed.head()
```

```python
natural_gas_production_transposed.drop(natural_gas_production_transposed.columns.difference(included_columns), 1, inplace=True)
natural_gas_production_transposed.dtypes
# natural_gas_production_transposed.plot(kind='line')
```

```python
renewable_share_df= pd.read_csv(renewables_electricity_share, encoding="ISO-8859-1")
renewable_share_df.head()
renewable_share_df.set_index('Country',inplace=True)
renewable_share_transposed=renewable_share_df.T
renewable_share_transposed.head()
```

```python
renewable_share_transposed.drop(renewable_share_transposed.columns.difference(included_columns), 1, inplace=True)
renewable_share_transposed = renewable_share_transposed.convert_objects(convert_numeric=True)
renewable_share_transposed.dtypes
renewable_share_transposed.plot(kind='line')
```
